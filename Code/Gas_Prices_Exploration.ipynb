# Annual API

### EIA Data Sets > Petroleum > Prices > Weekly Retail Gasoline and Diesel Prices > by Area > California > Annual

```python
# https://www.eia.gov/opendata/qb.php?category=240839&sdid=PET.EMM_EPM0_PTE_SCA_DPG.A

# Establishes my URL string
url_annual = 'http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=PET.EMM_EPM0_PTE_SCA_DPG.A'

# Inserts my API key
url_new_annual = url_annual.replace('YOUR_API_KEY_HERE', api_key)
```

```python
# Converts to json
api_response = requests.get(url_new_annual)
api_json_annual = api_response.json()
```

```python
# pprint(api_json_annual)
```

```python
# Navigates to key: series
api_json_series_a = api_json_annual['series']

# changes list back to dictionary
api_json_series_a = api_json_series_a[0]

# Navigates to key: data
api_json_data_a = api_json_series_a['data']
```

```python
# api_json_data_a
```

```python
# iterates through list prep for data frame
yearly_list = [['Year', 'Price']]
for i in api_json_data_a:
    yearly_list.append([i[0], i[1]])

# Creates DataFrame
yearly_df = pd.DataFrame(yearly_list[1:],columns=yearly_list[0])

# Drops Nan rows
yearly_df = yearly_df.dropna()
```

```python
yearly_df
```

```python
file_path = '..\Output\Annual_Prices.csv'

yearly_df.to_csv(file_path, index=False)
```

# Monthly API

### EIA Data Sets > Petroleum > Prices > Weekly Retail Gasoline and Diesel Prices > by Area > California > Monthly

```python
# https://www.eia.gov/opendata/qb.php?category=240839&sdid=PET.EMM_EPM0_PTE_SCA_DPG.A

# Establishes my URL string
url_month = 'http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=PET.EMM_EPM0_PTE_SCA_DPG.M'

# Inserts my API key
url_new_month = url_month.replace('YOUR_API_KEY_HERE', api_key)
```

```python
# Converts to json
api_response = requests.get(url_new_month)
api_json_month = api_response.json()
```

```python
# pprint(api_json_month)
```

```python
# Navigates to key: series
api_json_series_m = api_json_month['series']

# changes list back to dictionary
api_json_series_m = api_json_series_m[0]

# Navigates to key: data
api_json_data_m = api_json_series_m['data']
```

```python
# api_json_data_m
```

```python
# Preps a list for iteration
dates_list_m = [['Year', 'Month', 'Price']]

# Iterates through list and parses the date into separate year, month rows
for i in api_json_data_m:
    long_date = i[0]
    parsed_date = datetime.datetime.strptime(long_date, '%Y%m')
    year = parsed_date.date().year
    month = parsed_date.date().month
    dates_list_m.append([year, month, i[1]])

# Creates dataframe from list
monthly_df = pd.DataFrame(dates_list_m[1:],columns=dates_list_m[0])
```

```python
monthly_df
```

```python
file_path = '..\Output\Monthly_Prices.csv'

monthly_df.to_csv(file_path, index=False)
```

# Weekly API

### EIA Data Sets > Petroleum > Prices > Weekly Retail Gasoline and Diesel Prices > by Area > California > Weekly

```python
# https://www.eia.gov/opendata/qb.php?category=240839&sdid=PET.EMM_EPM0_PTE_SCA_DPG.A

# Establishes my URL string
url_week = 'http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=PET.EMM_EPM0_PTE_SCA_DPG.W'

# Inserts my API key
url_new_week = url_week.replace('YOUR_API_KEY_HERE', api_key)
```

```python
# Converts to json
api_response = requests.get(url_new_week)
api_json_week = api_response.json()
```

```python
# pprint(api_json_week)
```

```python
api_json_series_w = api_json_week['series']
api_json_series_w = api_json_series_w[0]
api_json_data_w = api_json_series_w['data']
```

```python
# api_json_data_w
```

```python
# Preps a list for iteration
dates_list_w = [['Year', 'Month', 'Day', 'Price']]

# Iterates through list and parses the date into separate year, month rows
for i in api_json_data_w:
        long_date = i[0]
        parsed_date = datetime.datetime.strptime(long_date, '%Y%m%d')
        year = parsed_date.date().year
        month = parsed_date.date().month
        day = parsed_date.date().day
        dates_list_w.append([year, month, day, i[1]])
        
# Creates dataframe from list
weekly_df = pd.DataFrame(dates_list_w[1:],columns=dates_list_w[0])
```

```python
weekly_df
```

```python
file_path = '..\Output\weekly_Prices.csv'

weekly_df.to_csv(file_path, index=False)
```
